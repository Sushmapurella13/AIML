```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

people=["kiran","Arun","vijay","Arun"]
age=[25,30,35,40]
height=[134,120,190,160]
weight=[45,56,34,78]
data={"name":people,"age":age,"height":height,"weight":weight}
df=pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['age'],df['weight'],color='blue',marker='o')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kg)')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(8,6))
plt.bar(df['name'],df['height'],color='green')
plt.title('Bar Chart of individuals\' Height')
plt.xlabel('Name')
plt.ylabel('Height(cm)')
plt.show()
```

```python
# 5. Histogram: Distribution of Age
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of age Distribution')
plt.xlabel('age (years)')
plt.ylabel('Frequency')
plt.show()
```
